```python
import niquests
from dotenv import load_dotenv
import base64
import os
from datetime import datetime, timedelta
from mypos.test import get_access_token
import json

load_dotenv()

client_id = os.getenv("MYPOS_CLIENT_ID")
client_secret = os.getenv("MYPOS_CLIENT_SECRET")
auth_base_url = os.getenv("MYPOS_AUTH_BASE_URL")
x_request_id = os.getenv("X_REQUEST_ID")
api_base_url = os.getenv("MYPOS_API_BASE_URL")
print(f"ENV: client_id: {client_id},\n client_secret: {client_secret},\n auth_base_url: {auth_base_url},\n x_request_id: {x_request_id},\n api_base_url: {api_base_url}")
print(f"Loaded successfully!!")

access_token = get_access_token()
print(f"Token: {access_token}")
```

```python
# CHeck transactions response

start_ts = datetime.now() - timedelta(days=1)
end_ts = datetime.now()

url = f"{api_base_url}/transactions"
headers = {
            "Authorization": f"Bearer {access_token}",
            "Content-Type": "application/json",
            "API-key": client_id,
            "X-Request-ID": x_request_id
        }

query_params = {
            "size": 100,
            "page": 1,
            "order": 0,
            "from_date": start_ts,
            "to_date": end_ts,
            "transaction_types" : "013"
        }
```

```python
response = niquests.get(url, headers=headers, params=query_params)
print(json.dumps(response.json(), indent=4))
```

```python
from typing import Optional, List

def get_multiple_transaction_details(api_base_url, access_token, client_id, x_request_id, payment_references: Optional[List[str]] = None):

    if payment_references is None:
        payment_references = ["POSA0192592GZQJ6"]

    url = f"{api_base_url}transactions/details"
    headers = {
        "Authorization": f"Bearer {access_token}",
        "Content-Type": "application/json",
        "API-key": client_id,
        "X-Request-ID": x_request_id
    }
    
    # Prepare query parameters with up to 5 payment references
    query_params = {
        "references": ",".join(payment_references[:5])  # Join up to 5 references with commas
    }

    try:
        response = niquests.get(url=url, headers=headers, params=query_params)
        response.raise_for_status()
        data = response.json()
        return data
       
        
    except niquests.exceptions.RequestException as e:
        print(f"Error fetching multiple transaction details: {e}")
        return {}

print(get_multiple_transaction_details(api_base_url, access_token, client_id, x_request_id, payment_references))
```
